```python
from netmiko import ConnectHandler # Test connection and send first command

cisco_ios = {
    'device_type': 'cisco_ios',
    'host':   'sandbox-iosxe-latest-1.cisco.com',
    'username': 'admin',
    'password': 'C1sco12345',
    'port' : 22,          # optional, defaults to 22
}

net_connect = ConnectHandler(**cisco_ios)
output = net_connect.send_command('show ip int brief')
print (output)
```

```python
line_str = output.splitlines() # Parse every line
print (line_str)
```

```python
line_str = output.splitlines() # show lines which you want
print (line_str[0])
print (line_str[1])
print (line_str[5])
print (line_str[8])
```

```python
# There is another way but a little bit complicated

for line in output.split('\n'):
    print(line)
```

```python
# Create list with this 'for loop'

String_list = []

for line in output.split('\n'):
    String_list.append(line)
print (String_list)



```

```python
# This both methods looks like give same result but not. There is a more advantage with using for loop. For example maybe you just want to have 'UP' interfaces.

for line in output.split('\n'):
    if 'up' in line:
        print (line)
```

```python
for line in output.split('\n'): # parse every 'up' interface line
    if 'up' in line:
        print (line.split())
```

```python
for line in output.split('\n'): # parse every 'up' interface line and use just interface name
    if 'up' in line:
        print (line.split()[0])
```

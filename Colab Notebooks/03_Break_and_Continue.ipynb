```python
# 'Break' stop the loop

Devices = ['RT01','RT02','RT03','SW01','SW02']

for dev in Devices:
    if dev == 'RT03':
        break
    print (dev)
print ('end of loop')
```

```python
# 'continue' bypass the item but not stop the loop 

Devices = ['RT01','RT02','RT03','SW01','SW02']

for dev in Devices:
    if dev == 'RT03':
        continue
    print (dev)
print ('end of loop')
```

```python
while True:
    num = int(input('Bir sayi giriniz :'))
    if num <= 10:
        continue
    if num > 20:
        break
    print (f'your number is {num}')
print ('End of loop') 
```

### Methods of Lists

```python
devices = ['RT01','RT02','SW01','SW02','FW01']

devices = devices.clear() # Clear method erase all value in list
print (devices) # None

devices = ['RT01','RT02','SW01','SW02','FW01']

# 'len' shows us that how many item in list
print (len(devices)) # 5

# 'count' count the specific items
print (devices.count('RT02')) # 1

# 'copy' copying list to another list
newDevices = devices.copy()
print (newDevices) # ['RT01', 'RT02', 'SW01', 'SW02', 'FW01']

# 'index' shows us the index of items. If there is more then one just show first one
print (devices.index('SW02')) # 3

# 'append' add value the end of list
devices.append('FW02')
print (devices) # ['RT01', 'RT02', 'SW01', 'SW02', 'FW01', 'FW02']

# 'insert' add value in specific index of list
devices.insert(0,'RT00')
print (devices) # ['RT00', 'RT01', 'RT02', 'SW01', 'SW02', 'FW01', 'FW02']

# 'extend' combine two list to one list

newDevices = ['SW03','SW04']
devices.extend(newDevices)
print (devices) # ['RT00', 'RT01', 'RT02', 'SW01', 'SW02', 'FW01', 'FW02', 'SW03', 'SW04']

# 'del' delete specific index of list
del devices[8]
print (devices) # ['RT00', 'RT01', 'RT02', 'SW01', 'SW02', 'FW01', 'FW02', 'SW03']

# 'pop' keep removed item

erasedItem = devices.pop(0)
print (devices) # ['RT01', 'RT02', 'SW01', 'SW02', 'FW01', 'FW02', 'SW03']
print (erasedItem) # RT00

# 'remove' erase value of item. You dont need to know its index

print (devices) # ['RT01', 'RT02', 'SW01', 'SW02', 'FW01', 'FW02', 'SW03']
devices.remove('SW02')
print (devices) # ['RT01', 'RT02', 'SW01', 'FW01', 'FW02', 'SW03']

# 'sort' sort the value of list but not keep another list

devices = ['RT02','RT01','SW01','SW02','FW01']
print (devices) # ['RT02', 'RT01', 'SW01', 'SW02', 'FW01']
devices.sort()
print (devices) # ['FW01', 'RT01', 'RT02', 'SW01', 'SW02']

# 'sorted' sort the value of list

devices = ['RT02','RT01','SW01','SW02','FW01']
newDevices = sorted(devices)
print (newDevices) # ['FW01', 'RT01', 'RT02', 'SW01', 'SW02']

# 'reverse' reverse of list

newDevices.reverse()
print (newDevices) # ['SW02', 'SW01', 'RT02', 'RT01', 'FW01']
```

## Dictionaries

Dictionaries are map structures based on `key` , `value` pairs. 

* Unordered
* Duplicated not allowed
* Changeable

```python
deviceDict = {
    'hostname':'RT01',
    'image':'IOS-XE',
    'ipAddr':'192.168.10.1'
}

print (deviceDict)
```

```python
deviceDict = {
    'hostname':'RT01',
    'image':'IOS-XE',
    'ipAddr':'192.168.10.1'
}

host = deviceDict['hostname'] # recall specific `value`
print(host)
```

```python
deviceDict = {
    'hostname':'RT01',
    'image':'IOS-XE',
    'ipAddr':'192.168.10.1'
}

keys = deviceDict.keys() # recall all keys
print (keys)
keylist = list(keys) # you can convert keys to list type
print (keylist)

values = deviceDict.values() # or values
print (values)
```

```python
deviceDict = {
    'hostname':'RT01',
    'image':'IOS-XE',
    'ipAddr':'192.168.10.1'
}

deviceDict['version'] = '17.0' # You can add key-value pair

print (deviceDict)
```

```python
deviceDict = {
    'hostname':'RT01',
    'image':'IOS-XE',
    'ipAddr':'192.168.10.1'
}

popVal = deviceDict.pop('ipAddr') # POP erase specific value but you can keep it
print (deviceDict) # {'hostname': 'RT01', 'image': 'IOS-XE'}
print (popVal) # 192.168.10.1
```

```python
deviceDict = {
    'hostname':'RT01',
    'image':'IOS-XE',
    'ipAddr':'192.168.10.1'
}

deviceDict.popitem() # popitem erase the last value of dict

print (deviceDict) # {'hostname': 'RT01', 'image': 'IOS-XE'}
```

```python
deviceDict = {
    'hostname':'RT01',
    'image':'IOS-XE',
    'ipAddr':'192.168.10.1'
}

del deviceDict # del delete the dictionary

print (deviceDict) # NameError: name 'deviceDict' is not defined
```

```python
deviceDict = {
    'hostname':'RT01',
    'image':'IOS-XE',
    'ipAddr':'192.168.10.1'
}

deviceDict.clear() # clear is erase content of dictionary
print (deviceDict) # {}
```

```python
deviceDict2 = dict( # Different expression of dictionary
    hostname = 'RT01',
    image = 'IOS-XE',
    ipAddr = '192.168.10.1'
)

print (deviceDict2) # {'hostname': 'RT01', 'image': 'IOS-XE', 'ipAddr': '192.168.10.1'}
```

## Tuples

* Ordered
* Unchangeable
* Allow dublicate

```python
devices = ('RT01','RT02','RT03','SW01','SW02')
print (type(devices)) # <class 'tuple'>
print (devices) # ('RT01', 'RT02', 'RT03', 'SW01', 'SW02')
print (devices[1]) # RT02
print (devices[-2]) # SW01
```

```python
devices = ('RT01','RT02','RT03','SW01','SW02')
newDevices = devices.append('SW03')
print (newDevices) # AttributeError: 'tuple' object has no attribute 'append'
```

```python
# But there is a workaround

devices = ('RT01','RT02','RT03','SW01','SW02')
devicesList = list(devices)
devicesList.append('SW03')
devices = tuple (devicesList)
print (devices)
```

```python
devices = ('RT01','RT02','RT03','SW01','SW02')
interfaces = ('Gi0/0/1',) # Can not add 'str' to 'tuple' but ',' sign convert interfaces str to tuple

devices += interfaces

print (devices)
```

```python
# Unpacking process

device = ('RT01', 'Gi0/0/1', '192.168.80.254')

(hostname,interface,ipAddress) = device

print (hostname)
print (interface)
print (ipAddress)
```

## Sets

* Unordered
* Changeable (mutable)
* Duplicates are not allowed

```python
setVar = {'RT01','RT02','SW01','SW02','RT01'}
print (setVar)
```

```python
devices = {'RT01','RT02','RT03','SW01','RT01'}

devices.add('RT04') # with 'add' method, you can add only one item

print (devices)
```

```python
devices = {'RT01','RT02','RT03','SW01'}

devices.update(['SW02','SW03']) # with 'update' method, you can add multiple items

print (devices)
```

### You can del item from SET with three ways:

* remove
* discard
* pop

```python
devices = {'RT01','RT02','RT03','SW01'}

devices.remove('SW01')
print(devices)

devices.remove('SW02') # If you try to del absent item, you will get an error 
```

```python
devices = {'RT01','RT02','RT03','SW01'}

devices.discard('SW01')
print (devices)

devices.discard('SW02') # It is same with 'remove' but you wont take an error
print (devices)
```

```python
devices = {'RT01','RT02','RT03','SW01'}

devices.clear() # it is clear the set but there is stil a set

print (devices)
```

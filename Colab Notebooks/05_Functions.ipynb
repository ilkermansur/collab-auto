# Functions

If you want to use code block againg and again, you should use 'function'.

```python
def firstFunc ():
    print ('this is first function')

firstFunc()
```

```python
# Functions insert between codes then continue after execution from this point

def firstFunc ():
    print ('this is function')

print ('this is 1')
print ('this is 2')
firstFunc()
print ('this is 3')
print ('this is 4')
firstFunc()
print ('this is 5')
```

```python
# You can use parameter 

def config_sw_port (interface,VlanID):
    cfgstr = f'''
Interface {interface}
Switchport mode access
Switchport access vlan {VlanID}
'''
    return cfgstr

print (config_sw_port ('gi0/1/0','10'))

# Interface gi0/1/0
# Switchport mode access
# Switchport access vlan 10

print (config_sw_port ('gi0/1/1','20'))

# Interface gi0/1/1
# Switchport mode access
# Switchport access vlan 20

# PS: You have to use all parameters.
```

## ARGS & KWARGS Usage

```python
# If you dont know that how many argument is insert to founction, you should use 'ARGS'

def sum (num1, num2):
    print (num1 + num2)
    
sum(14,15)

# lets use with 3 argument

sum (12,13,14)
```

```python
def sum (Val):
    print (type(Val))
sum(14) # type is 'int'
sum ('14') # type is 'str'
sum ({'hostname':'RT01'}) # type is 'dict'

```

```python
def sum (*Val): # If we use '*', type of this value is tuple allways. We store multiple value in 'tuple'
    print (type(Val))
sum(14) 
sum ('14') 
sum ({'hostname':'RT01'}) 
```

```python
def sum (*Val): 
    result = 0
    for item in Val:
        result += item
    print (result)
sum(14,17,19,10) 
```

## kwargs

If your values are dict, you should use 'kwargs'

```python
def sum (**Val): #
    print (type(Val))
sum() 
```

```python
def get_pers (**detail):
    for data in detail:
        print (data)

get_pers(username='cisco',password='Cisco1',port = 22)

```

```python
def get_pers (**detail):
    for data in detail.keys():
        print (data)

get_pers(username='cisco',password='Cisco1',port = 22)
```

```python
def get_pers (**detail):
    for data in detail.values():
        print (data)

get_pers(username='cisco',password='Cisco1',port = 22)
```

```python
def get_pers (**detail):
    for data in detail.items():
        print (data)

get_pers(username='cisco',password='Cisco1',port = 22)
```

```python
def get_pers (**detail):
    for key,val in detail.items():
        print (key,val)

get_pers(username='cisco',password='Cisco1',port = 22)
```

PS : The pass statement is used as a placeholder for future code. When the pass statement is executed, nothing happens, but you avoid getting an error when empty code is not allowed. Empty code is not allowed in loops, function definitions, class definitions, or in if statements.

```python

def passFunc ():
    pass

passFunc()
```
